### 3.3. FACES_DD 

FACES_DD jest zbiorem prywatnych zdjęć z lat 2003-2009. Zestaw zawiera 925 zdjęć **dużej rozdzielczości - średnio 6Mpix**, z oznaczonymi 2690 twarzami. 1205 twarzy oznaczone jest inicjałami dla 10 osób jak poniżej:

<br>AM': 11,
<br>'AS': 6,
<br>'BO': 24,
<br>'DD': 186,
<br>'KWD': 307,
<br>'OO': 27,
<br>'RO': 22,
<br>'TM': 10,
<br>'ZBD': 583,
<br>'ZD': 29.

Pozostałe 1485 oznaczone zostały jako 'Other'.<br>
Zdjęcia mogą zawierać o 0 do 84 oznaczonych twarzy.


<br>**Dla pretrenowanego MTCNN uzyskano AP 94,4%**
<br>Ze względu na unikalny charakter i zdjęcia dużej rozdzielczości zbiór może być wykorzystany do testów porównawczych modeli.
